```python
df_copy['Classes']=np.where(df_copy['Classes'].astype(str).str.contains('not fire'),'not fire','fire')
```

```python
print("Before Cleaning:")
print(df['Classes'].unique())
print(df['Classes'].value_counts())
```

```python
import numpy as np

# Step 1: Convert to Lowercase
df_copy['classes'] = df_copy['Classes'].astype(str)
print("After Conversion to String:", df_copy['Classes'].unique())  # Debugging
df_copy['Classes'] = df_copy['classes'].str.lower()
print("After Lowercasing:", df_copy['Classes'].unique())  # Debugging

# Step 2: Remove spaces
df_copy['Classes'] = df_copy['classes'].str.strip()
print("After Stripping Spaces:", df_copy['Classes'].unique())  # Debugging

# Step 3: Fix replacement
df_copy['Classes'] = df_copy['Classes'].replace({
    r'^\s*not\s*fire\s*$': 'not fire',
    r'^\s*fire\s*$': 'fire',
}, regex=True)
print("After Replacement:", df_copy['Classes'].unique())  # Debugging

# Final check
print("Final Counts:")
print(df_copy['Classes'].value_counts())
```
